```python
class File():
    def __init__(self):
        self.stockage = []
        
    def enfiler(self, v):
        self.stockage.insert(0, v)
        
    def defiler(self):
        return self.stockage.pop()
    
    def vide(self):
        return len(self.stockage)==0
```

```python
f = File()
f.enfiler(1)
f.enfiler(2)
f.enfiler(3)
```

```python
f.defiler()
```

```python
def parcours_largeur(g, depart):
    visite = { k:False for k in g.keys() }
    f = File()
    parcours = []
    
    f.enfiler(depart)
    
    while not f.vide():
        sommet = f.defiler()
        
        if visite[sommet]:
            continue
        
        visite[sommet] = True
        parcours.append(sommet)
        
        for v in g[sommet]:
            if not visite[v]:
                f.enfiler(v)
                
    return parcours
```

```python
p = parcours_largeur(graph, 'a')
```
